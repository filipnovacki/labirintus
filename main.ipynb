```python
from scripts.graph import DirGraph
from scripts.network import networker, plot_graph, plot_graph_circular,\
plot_graph_kk, plot_graph_planar
import networkx as nx
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
graph = DirGraph(['a', 'b', 'c', 'd', 'e', 'f'])

graph.add_edge('a', 'b', 0.3)
graph.add_edge('c', 'b', 0.5)
graph.add_edge('a', 'c', 0.2)
graph.add_edge('a', 'd', 0.7)
graph.add_edge('a', 'e', 0.6)
graph.add_edge('d', 'e', 0.4)
graph.add_edge('d', 'b', 0.1)
graph.add_edge('a', 'f', 0.8)
```

```python
G = networker(graph)
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color = "blue")
nx.draw_networkx_edges(G, pos, edgelist = G.edges())#, #edge_color=edge_colours)
nx.draw_networkx_labels(G, pos)
plt.show()
```
